```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from pmdarima import auto_arima

# Load your dataset (replace 'your_data.csv' with your file path and 'Electricity generation - TWh' with your column)

series = cleaned_df['Electricity generation - TWh']

# Step 1: Normalize the data
scaler = MinMaxScaler()
series_normalized = scaler.fit_transform(series.values.reshape(-1, 1)).flatten()

# Step 2: Use auto_arima to find optimal parameters
auto_model = auto_arima(series_normalized, 
                        start_p=0, start_q=0, max_p=5, max_q=5,
                        seasonal=False, trace=True,
                        error_action='ignore', suppress_warnings=True)

optimal_order = auto_model.order
print(f"Optimal ARIMA parameters: {optimal_order}")

# Step 3: Fit ARIMA model with optimal parameters
model = ARIMA(series_normalized, order=optimal_order)
results = model.fit()

# Step 4: Generate predictions
fitted_values_normalized = results.fittedvalues
fitted_values = scaler.inverse_transform(fitted_values_normalized.reshape(-1, 1)).flatten()

# Align actual values and predicted values
actual_values = series[1:]  # Skipping the first value due to differencing
predicted_values = fitted_values[:-1]

# Step 5: Calculate metrics
mae = mean_absolute_error(actual_values, predicted_values)
mse = mean_squared_error(actual_values, predicted_values)
r2 = r2_score(actual_values, predicted_values)

# Print metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")

# Step 6: Forecast future values
forecast = results.get_forecast(steps=12)
forecast_normalized = forecast.predicted_mean
forecast_values = scaler.inverse_transform(forecast_normalized.reshape(-1, 1))

print("Forecasted Values:")
print(forecast_values)

# Step 7: Residual analysis
residuals = actual_values - predicted_values
plt.figure(figsize=(10, 6))
plt.plot(residuals, label='Residuals')
plt.axhline(y=0, color='r', linestyle='--', label='Zero Line')
plt.title('Residual Analysis')
plt.legend()
plt.show()

# Step 8: Visualize actual vs predicted
plt.figure(figsize=(10, 6))
plt.plot(actual_values.values, label='Actual Values', color='blue')
plt.plot(predicted_values, label='Predicted Values', color='orange')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
```
